```python
import matplotlib.pylab as plt
import numpy as np
```

```python
np.log10(.5)/np.log10(.99)
```

```python
prices = np.linspace(2,100,9799)
```

```python
def price_adjustment(x):
    return round(x*.99,2)
```

```python
def price_check(original_price, current_price):
    if current_price >= .5*original_price:
        return True
    else:
        return False
```

```python
store = {}
for price in prices:
    price_list = []
    new_price = price_adjustment(price)
    while price_check(price, new_price):
        price_list.append(new_price)
        new_price = price_adjustment(new_price)
    price_list.append(new_price)
    store[round(price,2)] = len(price_list)
```

```python
lists = sorted(store.items()) # sorted by key, return a list of tuples

x, y = zip(*lists) # unpack a list of pairs into two tuples

plt.plot(x, y)
plt.show()
```
